# How to make a function in Python.

This is a simple Python function that adds two numbers.


```python
def add_func (a, b):
    result = a + b
    return result
```

1. def add_func(a, b):

def — keyword used to define a function.

add_func — the name of the function.

(a, b) — parameters (inputs the function receives).

2. result = a + b

Adds the values of a and b.

Stores the sum in a variable called result.

3. return result

Sends the calculated value back to where the function was called.

Example usage:

```python
add_func(1, 1)
```
